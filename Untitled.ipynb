```python
import os
import requests

# Create folder
folder = "images"
os.makedirs(folder, exist_ok=True)

# Download 100 images
for i in range(1, 101):
    # URL for random image (size 400×300 as an example)
    url = f"https://picsum.photos/400/300?random={i}"

    # File path
    filename = os.path.join(folder, f"dog_{i}.jpg")

    # Download and save
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, "wb") as f:
            f.write(response.content)
        print(f"Downloaded: {filename}")
    else:
        print(f"Failed to download image {i}")

print("✅ All images downloaded!")
```

```python
import cv2
import os

# Folders
input_folder = "images"
output_folder = "resized_images"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# List all files in the input folder
image_files = os.listdir(input_folder)

for filename in image_files:
    input_path = os.path.join(input_folder, filename)
    output_path = os.path.join(output_folder, filename)

    # Read image
    img = cv2.imread(input_path)

    if img is None:
        print(f"Skipping {filename} (cannot read)")
        continue

    # Calculate new dimensions (50% smaller)
    new_width = int(img.shape[1] * 0.5)
    new_height = int(img.shape[0] * 0.5)

    resized_img = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_AREA)

    # Save resized image
    cv2.imwrite(output_path, resized_img)
    print(f"Resized and saved: {output_path}")

print("✅ All images resized successfully!")
```
